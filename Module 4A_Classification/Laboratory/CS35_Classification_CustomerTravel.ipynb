```python
# read CSV file
import pandas as pd
from sklearn.preprocessing import LabelEncoder

Customertravel = pd.read_csv('CustomerTravel.csv')
```

```python
Customertravel.head() #top 5 records
```

```python
print(Customertravel.shape) #number of rows and columns
```

In Supervised Machine Learning such as Binary Classification, inputs are called **feature** values while outputs are called **label** values.

The codes below creates X array containing the feature values, and Y array containing the label values.

```python
X = Customertravel[["Age","FrequentFlyer", "AnnualIncomeClass", "ServicesOpted", "AccountSyncedToSocialMedia", "BookedHotelOrNot"]].values
Y = Customertravel["Target"].values

print(X.shape)
print(Y.shape)
```

```python
##print(Customertravel['AnnualIncomeClass'].value_counts())

# Identify columns with non-numerical data
non_numeric_cols = [col for col in ["FrequentFlyer", "AnnualIncomeClass", "ServicesOpted",
                                   "AccountSyncedToSocialMedia", "BookedHotelOrNot"]
                    if Customertravel[col].dtype == 'object'] # object is text or varchar

# Handle non-numerical columns (example using LabelEncoder)
label_encoders = {}
for col in non_numeric_cols:
    le = LabelEncoder()
    X[:, Customertravel.columns.get_loc(col)] = le.fit_transform(X[:, Customertravel.columns.get_loc(col)])
    label_encoders[col] = le  # Store encoders for later use if needed

X_df = pd.DataFrame(X, columns=["Age", "FrequentFlyer", "AnnualIncomeClass", "ServicesOpted",
                                "AccountSyncedToSocialMedia", "BookedHotelOrNot"])

print(X_df['AnnualIncomeClass'].value_counts())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_df, Y, test_size=0.25, random_state=42)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=8) #when choosing the number of neighbours, it is recommended to start with a smaller value, around 3 to 5

knn.fit(X_train, Y_train)
```

```python
Y_pred = knn.predict(X_test)

# Print the predictions
print("Predictions: {}".format(Y_pred))
```

```python
# Theory to be covered in next class

from sklearn.metrics import classification_report

report = classification_report(Y_test, Y_pred)

print(report)
```
