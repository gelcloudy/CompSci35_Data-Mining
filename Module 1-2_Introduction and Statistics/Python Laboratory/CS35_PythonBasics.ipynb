```python
# Defining recursive functions
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

num = int(input("Enter number: "))
print(f"Factorial of {num} is {factorial(num)}")
```

```python
# Classes and objects
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

    def display_info(self):
        print(f"'{self.title}' by {self.author}, published in {self.year}")

book01 = Book("1984", "George Orwell", 1949)
book02 = Book("Atomic Habits", "James Clear", 2018)
book02.display_info()
```

**Scipt below shows the calculation of the average age in years of the entire data set**
* A function age_years was created
* The function datetime.today was used, which is similar to the NOW() function in MS Excel
* The function strptime was also used, which converts a string into a date

```python
from datetime import datetime

def age_years(birthdate):
    return (datetime.today() - birthdate).days / 365.25 # TODAY() funcation of the DATETIME library is similar to the NOW() function of MS Excel

df['Birthdate_dt'] = df['Birthdate'].apply(datetime.strptime, args=("%Y-%m-%d",)) # Added the module name datetime before strptime and included the format string as an argument


print(df)



df['Age_years'] = df['Birthdate_dt'].apply(age_years)

avg_age = df['Age_years'].mean()

print(avg_age)



print(df.dtypes)
```
